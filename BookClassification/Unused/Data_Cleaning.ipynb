# Clean up information in Title section

## Remove any details after first slash mark (usually author/editor information).

```python
for i, row in df.iterrows():

    curr_title = row['Title']
    extract = re.search(r'(.+?(?=/))', curr_title)
    
    if extract is not None:
      df.at[i, 'Title'] = extract.group()
    else:
      df.at[i, 'Title'] = curr_title
```

```python
df.head()
```

```python
print(len(df))
```

## Remove punctuation

```python
df['Title'] = df['Title'].str.replace('[{}]'.format(string.punctuation), '')
```

```python
df['Notes'] = df['Notes'].str.replace('[{}]'.format(string.punctuation), '')
```

```python
df['Subjects'] = df['Subjects'].str.replace('[{}]'.format(string.punctuation), '')
df['Subjects'] = df['Subjects'].str.lower()
```

```python
df.head()
```

## Titles to Lower Case:

```python
df['Title'] = df['Title'].str.lower()
df['Subjects'] = df['Subjects'].str.lower()
df['Notes'] = df['Notes'].str.lower()
```

```python
df.head()
```

## Combine Subject headers and titles

```python
df['Combined'] = df['Title'] + " " + df['Subjects'] + " " + df['Notes']
df.head()
```
